# Project for My Mom

This is a quick project I did for my mom, who is a math teacher. Objective was to just show growth across the class. Pretty basic, but good exercise in Python.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

test = pd.read_csv('momdata.csv')
# dropping NA's
test.dropna(subset = ['pre1'], inplace = True)
#Numbering
test['code'] = np.arange(len(test)) + 1

print(test)

#Plot 1
plt.bar(test['code'] - 0.2, test['pre1'], 0.4, label = "Pre-Test")
plt.bar(test['code'] + 0.2, test['post1'], 0.4, label = "Post-Test")

plt.tick_params(
    axis='x',         
    which='both',      
    bottom=False,      
    top=False,         
    labelbottom=False)

plt.axhline(y=25, color='blue', linestyle='dotted')
plt.axhline(y=99, color='orange', linestyle='dotted')
plt.figtext(0.5, -0.1, "Dotted lines represent averages by test", wrap=True, horizontalalignment='center', fontsize=12)

plt.xlabel("By Student")
plt.ylabel("Test Score")
plt.title("Pre-Test and Post-Test Score, by Student, Concept 1")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol= 2)
plt.show()

plt.savefig('Concept1.png')
plt.clf()

# Plot 2
plt.bar(test['code'] - 0.2, test['pre2'], 0.4, label = "Pre-Test")
plt.bar(test['code'] + 0.2, test['post2'], 0.4, label = "Post-Test")

plt.tick_params(
    axis='x',         
    which='both',      
    bottom=False,      
    top=False,         
    labelbottom=False)

plt.axhline(y=10, color='blue', linestyle='dotted')
plt.axhline(y=91, color='orange', linestyle='dotted')
plt.figtext(0.5, -0.1, "Dotted lines represent averages by test", wrap=True, horizontalalignment='center', fontsize=12)

plt.xlabel("By Student")
plt.ylabel("Test Score")
plt.title("Pre-Test and Post-Test Score, by Student, Concept 2")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol= 2)
plt.show()

plt.savefig('Concept2.png')
plt.clf()

#Plot 3
plt.bar(test['code'] - 0.2, test['pre3'], 0.4, label = "Pre-Test")
plt.bar(test['code'] + 0.2, test['post3'], 0.4, label = "Post-Test")

plt.tick_params(
    axis='x',         
    which='both',      
    bottom=False,      
    top=False,         
    labelbottom=False)

plt.axhline(y=21, color='blue', linestyle='dotted')
plt.axhline(y=85, color='orange', linestyle='dotted')
plt.figtext(0.5, -0.1, "Dotted lines represent averages by test", wrap=True, horizontalalignment='center', fontsize=12)

plt.xlabel("By Student")
plt.ylabel("Test Score")
plt.title("Pre-Test and Post-Test Score, by Student, Concept 3")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol= 2)
plt.show()

plt.savefig('Concept3.png')
plt.clf()

x = ['Concept 1','Concept 2','Concept 3']
pre = [25,10,21]
post = [99,91,85]

x_axis = np.arange(len(x))

plt.bar(x_axis - 0.2, pre, 0.4, label = 'Pre-Test')
plt.bar(x_axis + 0.2, post, 0.4, label = 'Post-Test')
  
plt.xticks(x_axis, x)
plt.xlabel("Concepts")
plt.ylabel("Average Score")
plt.title("Average Test Score by Concept")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.25), ncol= 2)
plt.show()

plt.savefig('TAvg.jpeg')
plt.clf()
```
